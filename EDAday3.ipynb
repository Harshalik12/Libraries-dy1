```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Datawithabhishek/Google-Play-Store-Data-Analysis/refs/heads/main/Play%20Store%20Data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed:0'],inplace=True)
```

```python
df.head(2)
```

```python
cat_col=df.dtypes[df.dtypes=='object'].index
```

```python
cat_col
```

```python
for i in cat_col:
    plt.figure(figsize=(10,4))#creates the figure
    plt.title('Coutplot')#set a title as column name
    sns.countplot(x=df[i],palette='Set2')#seaborn plot
    plt.xticks(rotation=45)#rotate x_axis labels
    plt.tight_layout()#adjust layout
  
```
